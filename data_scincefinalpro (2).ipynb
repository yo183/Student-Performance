```python
gender_means = data_frame.groupby('gender')[['math_score', 'reading_score', 'writing_score', 'science_score']].mean()

plt.figure(figsize=(10, 6))
gender_means.T.plot(kind='bar', figsize=(10, 6), colormap='Set2')
plt.title('Average Scores by Gender', fontsize=16)
plt.xlabel('Subjects', fontsize=14)
plt.ylabel('Average Score', fontsize=14)
plt.legend(title='Gender')
plt.show
```
